```python
import matplotlib.pyplot as plt
import numpy as np

# Data extracted from the image
models = [
    'VQGAN', 'ADM', 'DiT-XL/2', 'RQ-Trans', 'BigGAN',
    'DiT(average)', 'GigaGAN', 'VAR-d16', 'DiT-XL/2-G(cfg=1.25)',
    'StyleGAN-XL', 'DiT-XL/2-G(cfg=1.50)', 'DiT-XL/2-ICE(Ours)'
]
fid_scores = [
    15.78, 10.94, 9.62, 7.55, 6.95,
    5.04, 3.45, 3.30, 3.22,
    2.30, 2.27, 2.26
]

# Index for "DiT-XL/2-ICE(Ours)"
our_model_index = models.index('DiT-XL/2-ICE(Ours)')

# --- Plotting ---
plt.style.use('seaborn-v0_8-whitegrid') # Using a seaborn style for a clean look

plt.figure(figsize=(14, 8)) # Increased figure size for better readability

# Plot the main line for all models
plt.plot(models, fid_scores, marker='o', linestyle='-', color='dodgerblue', markersize=8, linewidth=2, label='Other Models (FID-50k)')

# Highlight "DiT-XL/2-ICE(Ours)"
plt.plot(models[our_model_index], fid_scores[our_model_index],
         marker='o', color='red', markersize=12, markeredgewidth=2, markerfacecolor='darkred',
         label='DiT-XL/2-ICE(Ours)')

# --- Enhancements for Scientific Paper Standards ---

# Set title and labels with larger fonts
plt.title('Comparison of Image Generation Models (FID-50k Score)', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Model', fontsize=16, fontweight='bold', labelpad=15)
plt.ylabel('FID-50k Score ↓ (Lower is Better)', fontsize=16, fontweight='bold', labelpad=15)

# Rotate X-axis labels for better readability if model names are long
plt.xticks(rotation=45, ha='right', fontsize=12) # ha='right' aligns the rotated labels
plt.yticks(fontsize=12)

# Add data labels to each point for clarity
for i, score in enumerate(fid_scores):
    plt.text(models[i], score + 0.3, f'{score:.2f}', ha='center', va='bottom', fontsize=10, fontweight='normal')

# Add a specific annotation for "DiT-XL/2-ICE(Ours)"
our_model_name = models[our_model_index]
our_model_score = fid_scores[our_model_index]
plt.annotate(
    f'{our_model_name}\nFID: {our_model_score:.2f}',
    xy=(our_model_name, our_model_score),
    xytext=(our_model_name, our_model_score + 2.5), # Position the text above the point
    ha='center',
    fontsize=12,
    fontweight='bold',
    color='black',
    arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6),
    bbox=dict(boxstyle="round,pad=0.3", fc="yellow", ec="black", lw=1, alpha=0.8) # Added a bounding box
)

# Add a legend with larger font
plt.legend(fontsize=12, loc='upper right', frameon=True, shadow=True)

# Adjust layout to prevent labels from overlapping
plt.tight_layout(pad=1.5) # Add some padding

# Add a more prominent grid
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7) # Make grid lines more visible

# Set y-axis limits to give some space, similar to the original image
plt.ylim(0, max(fid_scores) + 2) # Start y-axis at 0 and add some padding at the top

# --- Display the plot ---
plt.show()
```
